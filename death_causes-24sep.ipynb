# Max Cause Each Year - WW

```python
death_max_year = []

for year in world['Year'].values:
  df_each_year = world[world['Year'] == year]
  max_value = float(df_each_year.max(axis=1).values)
  column = df_each_year[df_each_year == max_value].dropna(axis=1).columns.values[0]
  item = (column, year, max_value)

  death_max_year.append(item)

death_max_year
```
